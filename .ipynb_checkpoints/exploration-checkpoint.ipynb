Need to standardize the Pitcher Handedness to remove THROW_SIDE_KEY from context.
Attributes affected by handedness are,
- HORIZONTAL_BREAK
- RELEASE_SIDE
- HORIZONTAL_APPROACH_ANGLE

To handle this I will flip the sign for the above attributes for Left handed pitchers.


```python
data['THROW_SIDE_KEY'].value_counts()
```

```python
data[data['THROW_SIDE_KEY']=='R'].head()
```

```python
data[data['THROW_SIDE_KEY']=='L'].head()
```

```python
data['HORIZONTAL_BREAK'] = data.apply(lambda x: -x['HORIZONTAL_BREAK'] if x['THROW_SIDE_KEY'] == 'L' else x['HORIZONTAL_BREAK'], axis=1)
data['RELEASE_SIDE'] = data.apply(lambda x: -x['RELEASE_SIDE'] if x['THROW_SIDE_KEY'] == 'L' else x['RELEASE_SIDE'], axis=1)
data['HORIZONTAL_APPROACH_ANGLE'] = data.apply(lambda x: -x['HORIZONTAL_APPROACH_ANGLE'] if x['THROW_SIDE_KEY'] == 'L' else x['HORIZONTAL_APPROACH_ANGLE'], axis=1)
```

```python
data[data['THROW_SIDE_KEY']=='L'].head()
```

Just keeping independent variables (fixed effects) that affect Induced Vertical Break and Horizontal Break. For random effect we can take the Pitcher ID and maybe the pitch_type in context?

```python
data = data.loc[:,['PID', 'PITCHER_KEY', 'PITCH_TYPE_TRACKED_KEY', 'SPIN_RATE_ABSOLUTE',
       'RELEASE_SPEED', 'RELEASE_SIDE', 'RELEASE_HEIGHT', 'RELEASE_EXTENSION',
       'HORIZONTAL_APPROACH_ANGLE', 'VERTICAL_APPROACH_ANGLE','INDUCED_VERTICAL_BREAK', 'HORIZONTAL_BREAK']]
data
```

```python
model_vertical = smf.mixedlm("INDUCED_VERTICAL_BREAK ~ SPIN_RATE_ABSOLUTE + RELEASE_SPEED + RELEASE_SIDE + RELEASE_HEIGHT + RELEASE_EXTENSION + HORIZONTAL_APPROACH_ANGLE + VERTICAL_APPROACH_ANGLE", data, groups=data["PITCHER_KEY"])
result_vertical = model_vertical.fit()

data['residual_vertical'] = result_vertical.resid

# Calculate mean and standard deviation of residuals
mean_residual = np.mean(data['residual_vertical'])
std_residual = np.std(data['residual_vertical'])

# Determine threshold for "large" residuals
threshold = mean_residual + 2 * std_residual

# Calculate the proportion of residuals above this threshold
probability = np.sum(data['residual_vertical'] > threshold) / len(data)

print(f"Estimated probability that a pitch was affected by a dew point > 65Â°F: {probability:.2f}")

# Compute z-scores for each residual
data['z_score'] = (data['residual_vertical'] - mean_residual) / std_residual

# Compute the probability that each pitch was NOT affected by dew point > 65Â°F
data['prob_not_affected'] = data['z_score'].apply(norm.cdf)

# Compute the probability that each pitch WAS affected by dew point > 65Â°F
data['prob_affected_vertical'] = 1 - data['prob_not_affected']

print(data[['PID', 'prob_affected_vertical']])
```

```python
model_horizontal = smf.mixedlm("HORIZONTAL_BREAK ~ SPIN_RATE_ABSOLUTE + RELEASE_SPEED + RELEASE_SIDE + RELEASE_HEIGHT + RELEASE_EXTENSION + HORIZONTAL_APPROACH_ANGLE + VERTICAL_APPROACH_ANGLE", data, groups=data["PITCHER_KEY"])
result_horizontal = model_horizontal.fit()

data['residual_horizontal'] = result_horizontal.resid

# Calculate mean and standard deviation of residuals
mean_residual = np.mean(data['residual_horizontal'])
std_residual = np.std(data['residual_horizontal'])

# Determine threshold for "large" residuals
threshold = mean_residual + 2 * std_residual

# Calculate the proportion of residuals above this threshold
probability = np.sum(data['residual_horizontal'] > threshold) / len(data)

print(f"Estimated probability that a pitch was affected by a dew point > 65Â°F: {probability:.2f}")

# Compute z-scores for each residual
data['z_score'] = (data['residual_horizontal'] - mean_residual) / std_residual

# Compute the probability that each pitch was NOT affected by dew point > 65Â°F
data['prob_not_affected'] = data['z_score'].apply(norm.cdf)

# Compute the probability that each pitch WAS affected by dew point > 65Â°F
data['prob_affected_horizontal'] = 1 - data['prob_not_affected']

print(data[['PID', 'prob_affected_horizontal']])
```
